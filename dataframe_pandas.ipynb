**Dataframes usando pandas**

La libreria "Pandas" en Python nos sirve para analizar, manipular y explorar los datos.

**Funciones de Pandas**

*   Manejo eficiente de grandes cantidades de datos
*   Estructrras de datos intuitivas
*   Manipulación de datos
*   Análisis y estadísticas descriptivas
*   Visualización sencilla



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from random import randint
```

```python
alumnos = pd.read_csv("Alumnos.csv", encoding="latin-1")
print(alumnos)
```

```python
alumnos['Nombre Completo'] = alumnos['Nombre'] + ' ' + alumnos['Apellido']
print(alumnos)
```

```python
np.random.seed(42)
#numero de semillas que tieene un procesador de 64 bits -> np.random.seed(42)
```

```python
datos = {
    "Nombre Completo" : alumnos['Nombre Completo'],
    "Edad" : np.random.randint(18, 25, size=27),
    "Calificación" : np.random.normal(7.5, 1.0, size=27).round(1), #round(1) es para que aproxime un decimal
    "Asistencia (%)" : np.random.uniform(70, 100, size=27).round(2)  #round(2),
}

df = pd.DataFrame(datos)
print(df)
```

```python
plt.hist(df["Edad"], bins = 6, edgecolor = 'black', color = 'skyblue' )
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Distribución de Edades de los Alumnos")
plt.grid(axis='y', alpha= 0.7) #alpha para la opacidad
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df["Nombre Completo"], df["Edad"], color = 'lightgreen', edgecolor = 'black')
plt.xlabel("Alumno")
plt.ylabel("Edad")
plt.title("Edad de los Alumnos")
plt.xticks(rotation = 90)
plt.ylim(18, 25)
plt.grid(axis='y', linestyle = '--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df["Edad"], df["Asistencia (%)"], color = 'coral')
plt.xlabel("Edad")
plt.ylabel("Asistencia (%)")
plt.title("Relación entre Edad y Porcentaje de Asistencia")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df["Nombre Completo"], df["Calificación"], color = 'yellow', edgecolor = 'black')
plt.xlabel("Alumno")
plt.ylabel("Calificación")
plt.title("Calificaciones de los Alumnos")
plt.xticks(rotation = 90)
plt.ylim(5, 10)
plt.grid(axis='y', linestyle = '--', alpha=0.7)
plt.tight_layout()
plt.show()
```
